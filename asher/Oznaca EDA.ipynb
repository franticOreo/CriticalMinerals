# Hierarchical clustering


```python
import scipy
from scipy.cluster import hierarchy as sch
import sklearn
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df.columns
```

```python

df_scaled = df.drop(['Code','Wt Tot'], axis=1)
df_scaled = df_scaled[df_scaled != 'IS']

df_scaled = scaler.fit_transform(df_scaled)
df_scaled = pd.DataFrame(df_scaled)
df_scaled.columns = df.columns[2:]
df_scaled.head()
```

```python
df_scaled=df_scaled.dropna()
```

```python
plt.rcParams['figure.figsize'] = [15,12]
```

```python
dendro2 = sch.dendrogram(sch.linkage(df_scaled, method = 'complete'))
plt.title('Dendrogram')
plt.xlabel('Minerals')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')

cluster.fit(df_scaled) 
labels = cluster.labels_
labels
```

```python
len(labels)
```

```python
df_logscale = np.log(df_scaled)
sns.scatterplot('Ni', 'Zn', data=df_logscale, hue=labels)
```

```python
sns.scatterplot('Zn', 'Ga', data=df_logscale, hue=labels)
```
