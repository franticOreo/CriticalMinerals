````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

### Cleaning Convention

````python
df = pd.read_csv('WAMEX_1_Rockchip_Original.csv')
df.head()
````

````python
df_minerals = df.loc[:,df.columns.str.contains('ppm|ppb|pct')]
````

````python
df.isna().sum(axis=1)
````

````python
df_minerals.dtypes.unique()
````

````python
def geo_null_to_nan(val):
    if val == -9999 or val == -6666: 
        return np.nan
    else:
        return val
    
    
def abs_halve_neg_val(val):
#     print(val)
    if val < 0: 
        return abs(val) / 2
    else: 
        return val

df_minerals = df_minerals.applymap(geo_null_to_nan)
df_minerals = df_minerals.applymap(abs_halve_neg_val)
````

````python
df_minerals
````

### Unique Minerals from Sites with Zinc in Mindex

Reference for output from MINEDEX_EDA_COBALT.ipynb
```
zinc_minerals = ['Zinc', 'Silver', 'Copper', 'Lead', 'Gold', 'Cobalt']
```

````python
zinc_minerals = ['Zinc', 'Silver', 'Copper', 'Lead', 'Gold', 'Cobalt']
````

````python
full_to_atomic = {
    "Zinc": "Zn",
    "Silver": "Ag",
    "Copper": "Cu",
    "Lead": "Pb",
    "Gold": "Au",
    "Cobalt": "Co"
}

new_cols = []
for col in df_minerals.columns: 
    if "ppm" in col or "ppb" in col or "pct" in col:
        new_cols.append(col.split('_')[0])
    else:
        new_cols.append(col)
        
df_minerals.columns = new_cols
````

````python
minerals_atomic = [full_to_atomic[mineral] for mineral in zinc_minerals]
minerals_atomic
````

#### Remove Empty Cobalt Assays

````python
df_minerals = df_minerals[minerals_atomic]
df_minerals['lat'] = df['DLAT'] 
df_minerals['long'] = df['DLONG'] 
````

````python
df_w_Cu = df_minerals[~df_minerals['Cu'].isna()]
df_w_Cu.shape
````

````python
df_w_Cu_Pb = df_w_Cu[~df_w_Cu['Pb'].isna()]
df_w_Cu_Pb.shape
````
