#### Drop spurious values

````python
indexs_6666 = df[df.values == -6666].index

df = df.drop(index=indexs_6666)
````

#### Replace ```-9999``` Values

````python
def replace_null(row):
    return row.replace(-9999, 0)


df = df.apply(replace_null, axis=1)
````

````python
df.sample(5)
````

````python
df.shape
````

## Zinc

````python
df_has_zinc = df[df['Zn_ppm'] != 0]
````

#### Mapping

````python
df_has_zinc
````

````python
BBox = ((df_has_zinc.DLONG.min(),   df_has_zinc.DLONG.max(),      
         df_has_zinc.DLAT.min(), df_has_zinc.DLAT.max()))
BBox
````

````python
# map_img = plt.imread(image_path)

fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(df_has_zinc.DLONG, df_has_zinc.DLAT, zorder=1, alpha= 0.2, c='b', s=10)
ax.set_title('Plotting Spatial Data')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
# ax.imshow(map_img, zorder=0, extent = BBox, aspect= 'equal')
````

````python
df_has_zinc = df_has_zinc.filter(regex='ppm')
df_has_zinc.head(3)
````

````python
df_zinc_sums = df_has_zinc.sum().to_frame().sort_values(by=0, ascending=False)
df_zinc_sums[:10]
````

````python
# can't find Germaniun (Ge)
critical_minerals = ['In_ppm']
````

````python
df_has_zinc['In_ppm']
````

````python
df_zinc_sums[:5].plot(kind='bar')
````

````python
# top 10 highest sums of element ppm containing Zn
high_sum_el = df_zinc_sums[:10].index.to_list()
# Add Zn back in
high_sum_el = high_sum_el + ['Zn_ppm']
high_sum_el
````

````python
corr = df_has_zinc[high_sum_el].corr()
````

````python
plt.figure(figsize=(15,10))

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,
        cmap='summer')
````
