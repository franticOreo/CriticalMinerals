```python
import pandas as pd
import fiona
import geopy.distance
```

```python
SHAPE_PATH = '../data/minedex/MINEDEX_SHP_20210728/Minedex.shp'
```

```python
shape = fiona.open(SHAPE_PATH)
shape.next()['properties']
```

We have a bunch of OrderedDicts and we want to store them in a dataframe, how can we store 2

```python
len(next(iter(shape))['properties'])
```

```python
ordered_dicts = []

for item in shape.items():
    ordered_dicts.append(item[1])
```

```python
df = pd.DataFrame([ordered_dicts[i]['properties'] for i, _ in enumerate(ordered_dicts)])
df.head()
```

### Save Dataframe

```python
df.to_csv('../data/minedex/minedex_filtered_shp.csv')
```
